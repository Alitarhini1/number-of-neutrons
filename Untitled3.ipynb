```python
class Particle:
    def __init__(self, Z, A):
        self.Z=Z
        self.A=A
    def get_number_of_neutrons(self):
            return self.A-self.Z
        
example_particle=Particle(6,12)
number_of_neutrons=example_particle.get_number_of_neutrons()
print("Number of neutrons in the Particle: ",number_of_neutrons)
   
        
```
